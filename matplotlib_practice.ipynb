```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis,y_axis)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
ax = plt.axes()
ax.plot(x_axis,y_axis,marker = 'D',color = 'green', linewidth=3)
ax.set(title='PyBer')
```

```python
plt.plot(x_axis,y_axis,marker='*',color='blue',linewidth=2,label = 'Boston')
plt.xlabel('Date')
plt.ylabel('Fare($)')
plt.ylim(0,45)
plt.title('PyBer Fare by Month')
plt.legend()
```

```python
plt.barh(x_axis,y_axis,color='green',label='Boston')
plt.gca().invert_yaxis()
plt.xlabel('Date')
plt.ylabel('Fare($)')
plt.title('PyBer Fare By Month')
plt.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis,y_axis,'o')
```

```python
plt.scatter(y_axis,x_axis,color='red',marker='x',label='Chicago')
plt.legend()
plt.title('PyBer data v ride fare')
plt.xlim(0,45)
plt.gca().invert_yaxis()
```

```python
larger_y_values = []
for value in y_axis:
    larger_y_values.append(value*3)
plt.scatter(x_axis,y_axis,[i*3 for i in y_axis])
plt.gca().invert_xaxis()
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis,x_axis,s=[i*5 for i in y_axis],color='cyan',alpha=0.2,edgecolor='black',label='Chicago')
ax.set_xlabel('date')
ax.set_ylabel('Fare($) per ride')
ax.set_title('title')
ax.set_xlim(xmax=45)
ax.invert_yaxis()
ax.legend()
```
